# Промышленность: прогнозирование температуры стали

## Описание проекта

Металлургический комбинат **ООО "Так закаляем сталь"** имеет цель уменьшить потребление электроэнергии на этапе обработки стали, чтобы оптимизировать производственные расходы. Для достижения этой цели требуется построить модель, которая сможет предсказывать температуру стали на данном этапе.

*Этап обработки стали включает в себя несколько шагов:* 
1. Сначала сталь обрабатывается в металлическом ковше с вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, его внутренняя поверхность облицовывается огнеупорным кирпичом. Затем расплавленная сталь заливается в ковш и подогревается до нужной температуры при помощи графитовых электродов, установленных в крышке ковша.

2. На следующем этапе из стали выводится сера, производится десульфурация, корректируется химический состав путем добавления примесей, а также отбираются пробы для контроля качества. Далее, для изменения состава стали, она легируется путем подачи специальных добавок из бункера для сыпучих материалов или проволоки через трайб-аппарат.

3. Перед введением первых легирующих добавок производится измерение температуры стали и ее химический анализ. Затем температура повышается на несколько минут, добавляются легирующие материалы, и сплав продувается инертным газом. После этого производится перемешивание сплава и повторные измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

4. После этого расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки, откуда готовый продукт выходит в виде заготовок-слябов (плит). 

Построение модели, которая предсказывает температуру стали в этом процессе обработки, может помочь оптимизировать энергопотребление и улучшить эффективность производства на металлургическом комбинате.

## Цель

Цель проекта заключается в построении модели, способной предсказывать температуру стали на финальном этапе обработки, с целью уменьшения потребления электроэнергии и оптимизации производственных расходов на металлургическом комбинате.

#### Распределение мощностей

```python
_, axes = plt.subplots(2, 1, figsize=(16, 12))
data_arc_new[['start_of_arc_heating', 'active_power', 'reactive_power']].resample(
    '6H', on='start_of_arc_heating').mean().plot(ax=axes[0])
data_arc_new[['end_of_arc_heating', 'active_power', 'reactive_power']].resample(
    '6H', on='end_of_arc_heating').mean().plot(ax=axes[1])

axes[0].set_ylim(0.15, 1)
axes[1].set_ylim(0.15, 1)

axes[0].set_xlabel('Месяцы')
axes[1].set_xlabel('Месяцы')

axes[0].set_ylabel('Мощность')
axes[1].set_ylabel('Мощность')

axes[0].set_title('График мощностей в начале работы нагревателя', fontsize=14)
axes[1].set_title('График мощностей в конце работы нагревателя', fontsize=14)

plt.subplots_adjust(hspace=0.3)
plt.show()
```

Отметим, что графики довольно похожи друг на друга, однако на графиках выделяются странные зоны:
1. середина июля - вероятно был перерыв работе (c 2019-07-13 07:00:27 по 2019-07-18 06:04:18)
2. отрицательные значения по реактивной фазе в конце июля (их удалим, поскольку мощность не может быть отрицательной в процессе нагрева)

### Данные об электродах

```python
data_arc_new = data_arc_new.drop(data_arc_new[data_arc_new['reactive_power']<0].index).reset_index(drop=True)
```

**Сгенерируем новые факторы**

1. Расчитаем полную мощность, которая будет учитывать активную и реактивную мощности: $S = \sqrt{P^2 + Q^2}$ ,где P - активная мощность, Q - реактивная мощность

2. Посчитаем работу, как $N = T * S$, где T - время работы электрода, S - полная мощность

```python
# Полная мощность
data_arc_new['full_power'] = np.sqrt(data_arc_new['active_power'] + data_arc_new['reactive_power'])

# Удалим активную и реактивную мощность, поскольку эти данные учитываются в полной мощности
data_arc_new = data_arc_new.drop(columns=['active_power', 'reactive_power'])

# Посчитаем количество итераций нагрева в одной партии стали
data_arc_new['iterations'] = data_arc_new.groupby('key')['full_power'].transform('count')

# Добавим время работы электрода в секундах в итерации
data_arc_new['time_of_iteration'] = (data_arc_new['end_of_arc_heating'] - 
                                     data_arc_new['start_of_arc_heating']).dt.seconds

# Посчитаем работу
data_arc_new['work'] = data_arc_new['time_of_iteration'] * data_arc_new['full_power']

# Сгруппируем данные для дальнейшей работы
data_arc_new = data_arc_new.pivot_table(index='key', values=['full_power', 'work','iterations'],
                                        aggfunc='mean').reset_index()

data_arc_new.head()
```

```python
# посмотрим на распределение полученных признаков
_, axes = plt.subplots(2, 2, figsize=(12,12))

# построим гистограммы распределений признаков
data_arc_new['work'].plot(kind='hist', bins=50, ax=axes[0, 0])
data_arc_new['full_power'].plot(kind='hist', bins=50, ax=axes[0, 1])

# построим боксплоты
axes[1, 0].boxplot(data_arc_new['work'], vert=False)
axes[1, 1].boxplot(data_arc_new['full_power'], vert=False)

# подпишем оси
axes[0, 0].set_ylabel('Частота значений')
axes[0, 1].set_ylabel('Частота значений')
axes[0, 0].set_xlabel('Работа')
axes[0, 1].set_xlabel('Полная энергия')
axes[1, 0].set_xlabel('Работа')
axes[1, 1].set_xlabel('Полная энергия')

_.suptitle('Распределение признаков в данных об электродах', fontsize=14)
_.tight_layout(pad=1.9)

plt.show()

data_arc_new.describe().T
```

```python
# функция для удаления столбцов с содержанием пропусков > n%
def drop_empty_columns(data, percent):
    missing_percentages = data.isnull().mean()*100
    columns_to_drop = missing_percentages[missing_percentages > percent].index
    data = data.drop(columns=columns_to_drop)
    data = data.fillna(0)
    return data
```

```python
%%time
parameters = {
    'learning_rate': [0.01, 0.05],
    'iterations': [500, 1500]
}

catboost_model = CatBoostRegressor(random_state=RANDOM_SEED, logging_level='Silent')
catboost_grid_search = GridSearchCV(catboost_model, parameters, cv=CV, scoring='neg_mean_absolute_error')
catboost_grid_search.fit(X_train, y_train)

# Выведи лучшие параметры и оценку
print('Лучшие параметры: ', catboost_grid_search.best_params_)
print('Лучшая оценка: ', -catboost_grid_search.best_score_)
```

### Случайный лес

```python
%%time
rf_model = RandomForestRegressor(random_state=RANDOM_SEED)

parameters = {
    'n_estimators': [250, 700],
    'max_depth': [10, 15, 20]
}

rf_grid_search = GridSearchCV(rf_model, parameters, cv=CV, scoring='neg_mean_absolute_error', n_jobs=-1)
rf_grid_search.fit(X_train, y_train)

# Вывод лучших параметров и оценки
print('Лучшие параметры:', rf_grid_search.best_params_)
print('Лучшая оценка:', -rf_grid_search.best_score_)
```

На случайном лесе в процессе подбора гиперпараметов с кросс-валидацией равной 5 получили mae равное 6.84

**MAE**

Значение MAE (Mean Absolute Error) означает, что средняя абсолютная ошибка предсказания модели составляет 5.93. В данном случае, модель может отклоняться от фактического значения финальной температуры при выплавлении стали в среднем на 5.93 градуса. 


**Коэффициент детерминации** (R-квадрат) на тестовой выборке равный 0.38 означает, что модель CatBoost объясняет примерно 38% дисперсии зависимой переменной в тестовых данных. 

R-квадрат принимает значения от 0 до 1, где значение 1 означает, что модель идеально объясняет все вариации в данных, а значение 0 означает, что модель не объясняет никакую вариацию и прогнозы модели не лучше, чем простая медиана. Чем ближе значение R-квадрат к 1, тем лучше модель подстраивается под данные.

В нашем случае, значение R-квадрат равное 0.38 означает, что модель CatBoost объясняет 38% дисперсии в зависимой переменной на тестовой выборке. Таким образом, модель, хоть и объясняет только часть дисперсии, всё же предсказывает данные лучше, чем медиана.


```python
best_feature_importance = best_model.feature_importances_
feature_names = X_train.columns

catboost_importances = pd.Series(best_feature_importance, index=feature_names).sort_values(ascending=True)

fig, axes = plt.subplots(figsize=(16, 6))
catboost_importances.plot(kind='barh', ax=axes)
axes.set_title('Важность признаков модели CatBoost', fontsize=16)
axes.set_xlabel('Средняя важность признаков')
axes.set_ylabel('Название факторов')
fig.tight_layout()
```

Более высокие значения важности свидетельствуют о более значимом влиянии соответствующего признака на предсказания, в то время как более низкие значения важности указывают на менее влиятельные признаки.
